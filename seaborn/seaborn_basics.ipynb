# sns.set()

yeslae chai basicallly plot ko apperence lai ramro banai dinxa

The sns.set() function is used to set the aesthetic parameters in Seaborn, a popular Python visualization library based on Matplotlib. This function customizes the appearance of your plots, making them more attractive and informative by default.

Here's a brief overview of what sns.set() does and how to use it:

Syntax:


    import seaborn as sns

    sns.set(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
Parameters
context: Controls the scaling of plot elements. Options include 'paper', 'notebook', 'talk', and 'poster', in increasing order of size.
style: Sets the background and grid style. Options include 'white', 'dark', 'whitegrid', 'darkgrid', and 'ticks'.
palette: Sets the color palette. Options include built-in palettes such as 'deep', 'muted', 'bright', 'pastel', 'dark', and 'colorblind'.
font: Sets the font family. Options include any Matplotlib-supported font family.
font_scale: Controls the scaling of the font size. This is useful for making the text larger or smaller.
color_codes: If True, interprets shorthand color codes (e.g., b, g, r) in plotting functions.
rc: A dictionary of Matplotlib parameters to override Seaborn's default settings.

```python
#simply sns.set le chai plot ko appearance lai ramro banauxa
sns.set(style="darkgrid")
plt.plot(x,y)
plt.show()
```
