```python
cars=sns.load_dataset('mpg').dropna()
```

```python
sns.swarmplot(x=cars.origin,y=cars.horsepower)
```

# yellai violin plot mathi banayo vanae aajhai clear huinxa

```python
sns.swarmplot(x=cars.origin,y=cars.horsepower,size=3)
sns.violinplot(x=cars.origin,y=cars.horsepower)
#dot thulo vayae ra bhaira aako ho yo mathi ko size adjust garayo vanae hunxa
```
