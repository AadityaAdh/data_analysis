### here first and last rows are the same

```python
def deleter(df):

  for i in range(len(df)):
    if i not in range(len(df)):
      break
    for j in range(i+1,len(df)):
      if j not in range(len(df)):
        break
      if df.iloc[i,:].equals(df.iloc[j,:]):
        print(f"found duplicates in {i} ,{j}")
        df=df.drop(j,axis=0)



  return df
```

```python
deleter(dataframe1)
```

#pandas ko drop duplicate vanni function

    DataFrame.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)



Parameters
subset: Column label or sequence of labels, optional
Only consider certain columns for identifying duplicates. By default, all columns are used. vanya tyo column ma chai yeutai value xa vanae tyo row delete garni



keep: {'first', 'last', False}, default 'first'
Determines which duplicates (if any) to keep.
'first': Drop duplicates except for the first occurrence.
'last': Drop duplicates except for the last occurrence.
False: Drop all duplicates.



inplace: bool, default False
inplace=False (default): The operation returns a new DataFrame with the duplicates removed, leaving the original DataFrame unchanged.
inplace=True: The operation directly modifies the original DataFrame to remove the duplicates and returns None. This can save memory since no new DataFrame is created, but it means the original DataFrame is permanently changed.

```python
data={
    'name':['john','tom','tom','chris','harry','john'],
    'age':[23,24,25,26,27,23],
    'gender':['m','m','f','f','m','m'],
    'salary':[10000,20000,20000,40000,50000,10000]
}
dataframe2=pd.DataFrame(data)
dataframe2
```

```python
#deleting rows based on checking only the name column as both the name is same
#jaa jaa name same xa tyo sabbai delete garni
dataframe2.drop_duplicates(['name'],inplace=True)
dataframe2
```

```python
#deleting according to gender column yedi tyo column ma same data xa vayae tyo row nai delete
dataframe2.drop_duplicates(['gender'],inplace=True)
dataframe2
```
