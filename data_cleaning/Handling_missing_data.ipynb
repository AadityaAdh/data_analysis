#yesto bujam
data.loc[:,'name'] lae object return garxa jun lai loop ma chalauda tyo columnn vitra ko element aauxa. vanae xi yo use garni chai yedi single column lai matrai access garnu xa vanae


data.loc[:,["name"]]. chai dataframe ho aani data frame lai chai loop ma iterate garyo vanae column ko name aauxa aani yo aanusar chai element lai access garnu paryo

```python
def null_checker(df):
  dict={'null present in':''}
  for elements in df:
    for i in range(6):
      if df.loc[i,elements]==None:
        temp=dict['null present in']
        temp=temp+elements+','
        dict['null present in']=temp

  return dict
```

```python
null_values=null_checker(dataframe)
```

```python
print(null_values)
```

# 1 Lets do replacement first



### 1.1 Replacing with some default values

```python
def replacement_null(df,value):
  for column in dataframe:
    for i in range(6):
      if df.loc[i,column]==None:
        df.loc[i,column]=value

  return df
```

```python
df1=replacement_null(dataframe,"my default value")
```

```python
df1
```

#1.2 filling null values with previous values



```python
dataframe=pd.DataFrame(data)
```

```python
dataframe
```

### so here the none should be replaced by atul as it is the previous value

```python
def previous_filler(df):
  for column in dataframe:
    for i in range(6):
      if df.loc[i,column]==None:
        df.loc[i,column]=df.loc[i-1,column]

  return df

```

```python
df2=previous_filler(dataframe)
```

```python
df2
```

#1.4 filling null values with different values according to the column
eg salary column ko null value lai 10000 le replace garni aani name ko null vaue lai gopal lae replace garni

```python
dataframe1=dataframe
```

```python
def modifier(df,dict):
  for column in df:
    for i in range(6):
      if df.loc[i,column]==None:
        df.loc[i,column]=dict[column]

  return df
```

```python
dataframe1=modifier(dataframe1,{'name':'my default value'})
```

#1.5 filling with mean,min,max values



```python
data={
    'name':["aaditya","kunjan","puspan","atul","delta","aaditya "],
    'age':[21,22,23,24,25,26],
    'gender':["male","male","male","male","male","male"],
    'salary':[10000,20000,30000,40000,None,60000],
    'experience':[1,2,3,4,5,6]
}
```

```python
dataframe=pd.DataFrame(data)
dataframe1=dataframe.copy()
```

```python
dataframe1
```

## 1.5.1 filling with mean

```python
#paila suru yedi kunai column ko mean chainxa vanae
dataframe1.loc[:,['salary']].mean()
```

```python
#so aaba fillna ma chai yo value dini simple
dataframe1.fillna(value=dataframe1.loc[:,['salary']].mean())
```

# 1.5.2 filling with max value

```python
dataframe1.loc[:,['salary']].max()
```

```python
dataframe1=dataframe.copy()
dataframe1.fillna(value=dataframe1.loc[:,['salary']].max())
```

#1.5.3 filling with min value

```python
dataframe1=dataframe.copy()
dataframe1.fillna(value=dataframe1.loc[:,['salary']].min())
```
